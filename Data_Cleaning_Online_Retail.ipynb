```python
#Libraries
import pandas as pd
```

```python
#Data Import
df = pd.read_excel('/content/drive/MyDrive/Housing Data/online_retail/Online Retail.xlsx')
```

```python
#Data Overview
# To see the first few rows
df.head()
```

```python
# To see the summary statistics
df.describe(include='all')
```

```python
# To see the number of rows and columns
print("Shape of the data: ", df.shape)
```

```python
# To check for missing values in each column
df.isnull().sum()
```

```python
# Drop rows where 'CustomerID' is missing
df = df.dropna(subset=['CustomerID'])
```

```python
#Fill in the missing Description "Unknown"
df['Description'] = df.groupby('StockCode', group_keys=False)['Description'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
df.isnull().sum()
```

```python
# Check for negative values in 'Quantity' and 'UnitPrice'
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
# Convert 'InvoiceDate' to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Write DataFrame to an excel
df.to_excel('Cleaned_Online_Retail.xlsx', index=False)
```
